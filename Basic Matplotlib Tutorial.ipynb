## Plotting Higher Dimentional (>3D) Points
[Back to Index](#i)

#### Sample Data

```python
from sklearn.datasets import load_wine
x,y = load_wine(return_X_y=True)
print(x.shape," ",y.shape)
```

#### Reducing Data to 3D using PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=3)
pca_result = pca.fit_transform(x)

print(pca_result.shape)

x1 = pca_result[:,0]
x2 = pca_result[:,1] 
x3 = pca_result[:,2]
```

#### Plotting Reduced 3D Data

```python
import matplotlib.pyplot as plt
import matplotlib
import matplotlib.patches as mpatches
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = Axes3D(fig)
colors = ['red','green','blue']

ax.scatter(x1,x2,x3,c=y,cmap=matplotlib.colors.ListedColormap(colors))

red_patch = mpatches.Patch(color='red', label='Class 0')
green_patch = mpatches.Patch(color='green', label='Class 1')
blue_patch = mpatches.Patch(color='blue', label='Class 2')
plt.legend(handles=[red_patch,green_patch,blue_patch])

plt.show()
```

#### Reducing Data to 2D using PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(x)

print(pca_result.shape)

x1 = pca_result[:,0]
x2 = pca_result[:,1]
```

#### Plotting Reduced 2D Data

```python
import matplotlib.pyplot as plt
import matplotlib.patches as mpatches
import matplotlib

fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
#the dataset has 3 Classes
colors = ['red','green','blue']

ax.scatter(x1,x2,c=y,cmap=matplotlib.colors.ListedColormap(colors))

red_patch = mpatches.Patch(color='red', label='Class 0')
green_patch = mpatches.Patch(color='green', label='Class 1')
blue_patch = mpatches.Patch(color='blue', label='Class 2')
plt.legend(handles=[red_patch,green_patch,blue_patch])

plt.show()
```

### Plotting a single line

```python
import matplotlib.pyplot as plt

plt.title("Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

x = [45,19,20,12,22]
y = [12,12,23,35,54]

#matplotlib plots the points in the same left to right order as mentioned in the list 'x' and 'y'
plt.plot(x,y)
plt.show()
```

### Plotting multiple lines in the same graph

```python
import matplotlib.pyplot as plt

plt.title("Graph")
plt.xlabel("X axis")
plt.ylabel("Y axis")

x = [12,19,20,12,22]
y = [12,12,23,35,54]
plt.plot(x,y, label = '1')

x = [1,12,23,14,25]
y = [12,12,23,35,54]
plt.plot(x,y, label = '2')

plt.legend()
plt.show()
```

#### Subplots in a plot

```python
import glob
from PIL import Image

def plot_grid(data2,grid):
    plt.title("Output")
    for i in range(1, grid*grid+1):
        plt.subplot(grid, grid, i)
        plt.imshow(data2[i-1])
        plt.xticks([])
        plt.yticks([])
    #add padding
    plt.tight_layout()
    plt.show()
    
def read_all_images(path):
    data2 = []
    for img in glob.glob(path+"/*.jpg"):
        data2.append(Image.open(img))
    return data2

data2=read_all_images("cat")
grid=2 # we need a 2 x 2 grid
plot_grid(data2,grid)
```
